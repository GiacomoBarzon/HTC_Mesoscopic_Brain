```python
from HTC import HTC

import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# Get network names
folder = 'results/random/'
names = os.listdir(folder)
names = [x.rsplit('_', 2)[0] for x in names if not 'pdf' in x]
names = np.unique(names)

names
```

```python
names = names[-1]
names
```

```python
N = 100
fully = []

for i in range(N):
    fully.append( HTC.loadFromName(folder + names+'_'+str(i)) )
```

```python
fig, axs = plt.subplots(3, 2, figsize=(15,15))
tw = [x.twinx() for row in axs for x in row]

alpha = 0.1

# Plot everything
for i in range(N):
    axs[0,0].plot(fully[i].Trange/fully[i].Tc, fully[i].A, 'b', alpha=alpha)
    tw[0].plot(fully[i].Trange/fully[i].Tc, fully[i].sigmaA, 'g', alpha=alpha)
    
    axs[0,1].plot(fully[i].Trange/fully[i].Tc, fully[i].A_norm, 'b', alpha=alpha)
    tw[1].plot(fully[i].Trange/fully[i].Tc, fully[i].sigmaA_norm, 'g', alpha=alpha)
    
    axs[1,0].plot(fully[i].Trange/fully[i].Tc, fully[i].S1, 'b', alpha=alpha)
    tw[2].plot(fully[i].Trange/fully[i].Tc, fully[i].S2, 'g', alpha=alpha)
    axs[1,1].plot(fully[i].Trange/fully[i].Tc, fully[i].S1_norm, 'b', alpha=alpha)
    tw[3].plot(fully[i].Trange/fully[i].Tc, fully[i].S2_norm, 'g', alpha=alpha)
    
    axs[2,0].plot(fully[i].Trange/fully[i].Tc, fully[i].Fisher, 'b', alpha=alpha)
    #tw[4].plot(random[0][i].Trange/random[0][i].Tc, random[0][i].sigmaC, 'g', alpha=alpha)
    axs[2,1].plot(fully[i].Trange/fully[i].Tc, fully[i].Fisher_norm, 'b', alpha=alpha)
    #tw[5].plot(random[0][i].Trange/random[0][i].Tc, random[0][i].sigmaC_norm, 'g', alpha=alpha)

# Set ylim and ylabel
axs[0,0].set_ylim([0.01, 0.3])
axs[0,1].set_ylim([0.01, 0.3])
tw[0].set_ylim([0.01, 0.12])
tw[1].set_ylim([0.01, 0.12])
axs[0,0].set_ylabel('A', fontsize=13)
tw[1].set_ylabel(r'$\sigma_A$', fontsize=13)

axs[1,0].set_ylim([0.018, 0.27])
axs[1,1].set_ylim([0.018, 0.27])
tw[2].set_ylim([0.0151, 0.015175])
tw[3].set_ylim([0.0151, 0.015175])
axs[1,0].set_ylabel('S1', fontsize=13)
tw[3].set_ylabel('S2', fontsize=13)

axs[2,0].set_ylim([0,0.014])
axs[2,1].set_ylim([0,0.014])
axs[2,0].set_ylabel('Fisher', fontsize=13)
axs[2,0].set_xlabel('T/Tc', fontsize=13)
axs[2,1].set_xlabel('T/Tc', fontsize=13)

axs[0,0].set_title('W', fontsize=13)
axs[0,1].set_title(r'$W_{norm}$', fontsize=13)

# Set grid
for row in axs:
    for ax in row:
        ax.grid()
        
plt.show()
```
